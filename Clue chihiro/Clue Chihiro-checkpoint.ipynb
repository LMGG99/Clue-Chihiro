```python
from os import putenv
from tkinter import *  #Importa todos los modulos de la libreria

root = Tk() 
root.title("Clue - Chihiro Edition")
root.geometry("800x780") #Define el ancho y alto de la ventana
root.resizable(0,0)

bg = PhotoImage(file = "mapa.png")                           #Carga de imagenes
BarraMenu=PhotoImage(file = "barra menu.jpg")
imagen_Dormitorios = PhotoImage(file = "Dormitorios.png")
imagen_kamshi = PhotoImage(file = "kamshi trabajo.png")
imagen_casazeniba = PhotoImage(file = "casa zeniba.png")
imagen_oficinayubaba = PhotoImage(file = "oficina yubaba.png")
imagen_puente = PhotoImage(file = "puente.png")

haku_Img = PhotoImage(file = "haku.png")
noface_Img = PhotoImage(file = "noface.png")
chihiro_Img = PhotoImage(file = "chihiro.png")
bb_Img = PhotoImage(file = "bb.png")
papas_Img = PhotoImage(file = "papas.png")

sapo_Img = PhotoImage(file = "sapo.png")
lin_Img = PhotoImage(file = "lin.png")
yubaba_Img = PhotoImage(file = "yubaba.png")
zeniba_Img = PhotoImage(file = "zeniba.png")
kamashi_Img = PhotoImage(file = "kamashi.png")

BarraDialogo = PhotoImage(file = "barra dialogo.png")


import random

nombres = ['Sapo','Lin','Yubaba','Zeniba','Kamashi']
images =[sapo_Img,lin_Img,yubaba_Img,zeniba_Img,kamashi_Img]
lugar = ['Puente', 'Oficina Yubaba','Casa Zeniba','Trabajo Kamashi','Dormitorios']
pista = ['Trozo de carbon','Garra Mojada','Hilo suelto','Raton','Trozo de oro']

aa=random.randint(0,4)
Malvado =[nombres[aa],lugar[random.randint(0,4)],pista[random.randint(0,4)],images[aa]]


Mapa=[] 
Conclusion=[]
AccionesCount=0 
for i in range(5):
    aa=random.randint(0,(len(nombres)-1))
    Mapa.append([nombres[aa],lugar[random.randint(0,len(lugar)-1)],pista[random.randint(0,len(pista)-1)],images[aa]])
    nombres.remove(Mapa[i][0])
    lugar.remove(Mapa[i][1])
    pista.remove(Mapa[i][2])
    images.remove(Mapa[i][3])


canvas1 = Canvas( root, width = 800, 
                        height = 780) 

canvas1.pack(fill = "both", expand = True) 
canvas1.pack() 

canvas1.create_image( 0, 0, image = bg, anchor = "nw")
canvas1.create_image(0,0,image=BarraMenu,anchor="nw")

canvas1.create_text( 400, 90, 
                    text="Elija un lugar para investigar", fill="Black",
                    font=("Helvetica",18)) 
canvas1.create_text( 400, 120, 
                    text=f"Intervenciones Restantes: {5-AccionesCount}", fill="Black",
                    font=("Helvetica",18)) 


def cambia_Dormitorios():
    canvas1.pack(fill = "both", expand = True) 
    canvas1.pack() 
    canvas1.create_image( 0, 0, image = imagen_Dormitorios, anchor = "nw")
    ocultar_botones()
    global nom_lugar
    nom_lugar="Dormitorios"
    global i_mapa
    i_mapa=ubicar_mapa(nom_lugar)
    Destino(nom_lugar)
    
def cambia_Kamshi():
    canvas1.pack(fill = "both", expand = True) 
    canvas1.pack() 
    canvas1.create_image( 0, 0, image = imagen_kamshi, anchor = "nw")
    ocultar_botones()
    global nom_lugar
    nom_lugar="Trabajo de Kamashi"
    global i_mapa 
    i_mapa=ubicar_mapa(nom_lugar)
    Destino(nom_lugar)

    
def cambia_puente():
    canvas1.pack(fill = "both", expand = True) 
    canvas1.pack() 
    canvas1.create_image( 0, 0, image = imagen_puente, anchor = "nw")
    ocultar_botones()
    global nom_lugar
    nom_lugar="Puente"
    global i_mapa
    i_mapa=ubicar_mapa(nom_lugar)
    Destino(nom_lugar)
    
def cambia_oficina_yubaba():
    canvas1.pack(fill = "both", expand = True) 
    canvas1.pack() 
    canvas1.create_image( 0, 0, image = imagen_oficinayubaba, anchor = "nw")
    ocultar_botones()
    global nom_lugar
    nom_lugar="Oficina Yubaba"
    global i_mapa
    i_mapa=ubicar_mapa(nom_lugar)
    Destino(nom_lugar)
    
def cambia_casa_zeniba():
    canvas1.pack(fill = "both", expand = True) 
    canvas1.pack() 
    canvas1.create_image( 0, 0, image = imagen_casazeniba, anchor = "nw")
    ocultar_botones()
    global nom_lugar
    nom_lugar="Casa Zeniba"
    global i_mapa
    i_mapa=ubicar_mapa(nom_lugar)
    Destino(nom_lugar)
    
#--------------------------------------------- Mostrar/Ocultar botones del mapa    
def ocultar_botones():
    Oficina_Yubaba.place_forget()
    Kamshi.place_forget()
    Dormitorios.place_forget()
    casa_zeniba.place_forget()
    puente.place_forget()
    
    global EnLugar
    global DialogCount
    EnLugar=1
    DialogCount=0
    
def mostrar_botones():
    boton_investigar.place_forget()
    boton_preguntar.place_forget()
    Oficina_Yubaba.place(x=90,y=315)
    Kamshi.place(x=120,y=540)
    Dormitorios.place(x=330,y=315)
    casa_zeniba.place(x=400,y=695)
    puente.place(x=450,y=450)
    
def mostrar_opciones():
    boton_uno.place(x=400,y=300)
    boton_dos.place(x=400,y=340)
    boton_tres.place(x=400,y=380)
    boton_cuatro.place(x=400,y=420)
    boton_cinco.place(x=400,y=460)

#----------------------------------------------------- Dialogo general del destino
def Destino(lugar):
    
    texto=["Haku:\nMuy bien chicos, llegamos a  "+lugar,
           "No face: \nY al parecer no somos los unicos. \n"+Mapa[i_mapa][0]+" aqui tambien está",
           "Haku: \nEntonces, que hacemos?"]
    ImgTex=[haku_Img,noface_Img,haku_Img]
    
    global dialogo
    global Imagen

    if DialogCount>0 and DialogCount<len(texto):  

        canvas1.itemconfig(Imagen,image=ImgTex[DialogCount])
        canvas1.itemconfig(dialogo,text=texto[DialogCount])
        if DialogCount==2:
            boton_investigar.place(x=450,y=320)
            boton_preguntar.place(x=450,y=400)
        
    elif DialogCount>=len(texto): 
        siguiente.place_forget()
        
        
    else:             
        Imagen=canvas1.create_image( 0, 0, image = haku_Img, anchor = "nw")
        dialogo=canvas1.create_text( 30, 640, fill="Black",
                        text=texto[DialogCount], anchor="nw",
                        font=("Helvetica",16))
        
def conversar():

    boton_investigar.place_forget()
    boton_preguntar.place_forget()
    siguiente.place(x=700,y=550)
    global AccionesCount
    global DialogCount
    global EnLugar
    global num_dialogo 
    global dialogo  
    global Imagen
    
    canvas1.delete(Imagen)
    canvas1.delete(dialogo)
    AccionesCount+=1
    EnLugar=4
    DialogCount=0
    boton_menu.place_forget()
    num_dialogo=random.randint(0,2) 
    HablarSospechoso()
    
def investigar():
    boton_investigar.place_forget()
    boton_preguntar.place_forget()
    siguiente.place(x=700,y=550)
    global AccionesCount
    global EnLugar
    global DialogCount
    global dialogo  
    global Imagen
    
    canvas1.delete(Imagen)
    canvas1.delete(dialogo)
    AccionesCount+=1
    EnLugar=2
    DialogCount=0
    boton_menu.place_forget()
    EncontrarPista()

def HablarSospechoso():
    global dialogo
    global Imagen
    
    lugRandom=random.randint(0,4)
    nomRandom=random.randint(0,4)
    if Mapa[nomRandom][0]==Mapa[i_mapa][0] or Mapa[nomRandom][0]==Malvado[0]: 
        nomRandom=random.randint(0,4)
    if lugRandom==i_mapa:
        lugRandom=random.randint(0,4)
        
    dialogo1=[f"No face: \nHola {Mapa[i_mapa][0]}.",   #Dialogo cuando se habla con el malvado
             f"{Mapa[i_mapa][0]}: \nOh, hola chicos, en qué puedo ayudar?",
             f"BB: \nNos preguntabamos donde te encontrabas durante las ultimas horas",
             f"{Mapa[i_mapa][0]}: \nMmm.., estaba en {Mapa[lugRandom][1]}, y recuerdo haber visto a {Mapa[nomRandom][0]} por ahí.\nDe hecho, se veia muy sospechoso y quiza dejó alguna pista ahi",
             f"No face: \nMmmm interesante ... Muchas gracias por tu ayuda, {Mapa[i_mapa][0]}",
             f"{Mapa[i_mapa][0]}: \nNo fue nada, cuidense..."]
    image=[noface_Img,Mapa[i_mapa][3],bb_Img,Mapa[i_mapa][3],noface_Img,Mapa[i_mapa][3]]
    #--------- Dialogos de sospechosos inocentes --------------
    dialogos=[[f"{Mapa[i_mapa][0]}: \nHola chicos ... pasó algo?",
                 "Chihiro: \nHola... alguien secuestró a mis padres y estamos buscandolos.\nNos podria decir donde ha estado la ultimas horas?",
                 f"{Mapa[i_mapa][0]}: \nOoh, demonios. espero lo encuentren pronto. \nEstuve en los baños tomando un descanso, de hecho, ahora que lo mencionas, ahí dejé mis llaves; espero sigan ahi.",
                 "Chihiro: \nMis padres no pudieron haber estado ahi\nEllos buscaban comida",
                 "Haku: \nMuchas gracias por la informacion, y espero encuentre sus llaves",
                 f"{Mapa[i_mapa][0]}: \nMuchas gracias, y espero encuentren a tus padres chihiro"],
            [f"Haku: \nQue tal, {Mapa[i_mapa][0]}?",
                 f"{Mapa[i_mapa][0]}: \nHola chicos",
                 f"No face: \nAlguien secuestró a los padres de chihiro, y las luces de ciertas zonas estuvieron fallando, \npor lo que no podemos saber quien se los llevo. \nNos podria decir donde estuviste estas horas?",
                 f"{Mapa[i_mapa][0]}: \nEstaba en el puente con un viejo amigo de la infancia. \nSi no me crees puedes preguntar.",
                 "No face: \nTe creemos, las luces sí estaban funcionando ahi, \nasi que el responsable no estuvo por ahí.",
                 f"{Mapa[i_mapa][0]}: \nBueno, yo me retiro, esos baños herbales me está esperando."],
            [f"{Mapa[i_mapa][0]}: \nHola chicos, en que les puedo ayudar?",
                 f"BB: \nHola {Mapa[i_mapa][0]}, disculpe, nos podria ayudar diciendo donde ha estado las ultimas horas?",
                 f"{Mapa[i_mapa][0]}:\nMmm... dejen recordar ...",
                 f"{Mapa[i_mapa][0]}:\nEstaba con {Mapa[nomRandom][0]} en la cocina, \ny despues yo me dirigí a mi cuarto",
                 f"No face: \nEntiendo...gracias, {Mapa[i_mapa][0]}"]]
    image2 =[[Mapa[i_mapa][3],chihiro_Img,Mapa[i_mapa][3],chihiro_Img,haku_Img,Mapa[i_mapa][3]],
           [haku_Img,Mapa[i_mapa][3],noface_Img,Mapa[i_mapa][3],noface_Img,Mapa[i_mapa][3]],
           [Mapa[i_mapa][3],bb_Img,Mapa[i_mapa][3],Mapa[i_mapa][3],noface_Img]]
    if Malvado[0]==Mapa[i_mapa][0]: 
        if DialogCount>0 and DialogCount<len(dialogo1):  
            canvas1.itemconfig(Imagen,image =image[DialogCount])
            canvas1.itemconfig(dialogo,text=dialogo1[DialogCount])
        
        elif DialogCount>=len(dialogo1): 
            
            volver_menu()
            
        
        else:            
            Imagen=canvas1.create_image( 0, 0, image = image[DialogCount], anchor = "nw")
            dialogo=canvas1.create_text( 30, 640, fill="Black",
                        text=dialogo1[DialogCount], anchor="nw",
                        font=("Helvetica",16))
    else:   
        if DialogCount>0 and DialogCount<len(dialogos[num_dialogo]):  
            canvas1.itemconfig(Imagen,image =image2[num_dialogo][DialogCount])
            canvas1.itemconfig(dialogo,text=dialogos[num_dialogo][DialogCount])
        
        elif DialogCount>=len(dialogos[num_dialogo]):
            
            volver_menu()
           
        
        else:      
            Imagen=canvas1.create_image( 0, 0, image = image2[num_dialogo][DialogCount], anchor = "nw")
            dialogo=canvas1.create_text( 30, 640, fill="Black",
                        text=dialogos[num_dialogo][DialogCount], anchor="nw",
                        font=("Helvetica",16))
    
    
def EncontrarCollar():   
    global dialogo
    global Imagen
    
    dialogo1=["Chihiro: \nChicos, encontré la cartera de mi papá. Aqui fue donde estuvieron.",  
            "BB: \nPobres, debe estar asustados donde sea que estén.",
             "Chihiro: \nNo me dentendré hasta encontrarlos.",
             "Haku: \nDescuida, Chihiro. Vamos a encontrarlos."]
    ImgTex1=[chihiro_Img,bb_Img,chihiro_Img,haku_Img]
    
    dialogo2=["No face: \nMmm... no hay nada aqui que indique que estuvieran aqui", 
             "Haku: \nTal vez no estuvieron por aquí",
             "Chihiro: \nDonde podran estar?..."]
    ImgTex2=[noface_Img,haku_Img,chihiro_Img]
    
    if Malvado[1]==Mapa[i_mapa][1]:
        if DialogCount>0 and DialogCount<len(dialogo1):  
            canvas1.itemconfig(Imagen,image =ImgTex1[DialogCount])
            canvas1.itemconfig(dialogo,text=dialogo1[DialogCount])
            
        elif DialogCount>=len(dialogo1):
            
            volver_menu()
            
        
        else:            
            Imagen=canvas1.create_image( 0, 0, image = ImgTex1[DialogCount], anchor = "nw")
            dialogo=canvas1.create_text( 30, 640, fill="Black",
                        text=dialogo1[DialogCount], anchor="nw",
                        font=("Helvetica",16))
    else:  
        if DialogCount>0 and DialogCount<len(dialogo2):  
            canvas1.itemconfig(Imagen,image =ImgTex2[DialogCount])
            canvas1.itemconfig(dialogo,text=dialogo2[DialogCount])
    
        elif DialogCount>=len(dialogo2): 
            
            volver_menu()
            
        
        else:         
            Imagen=canvas1.create_image( 0, 0, image = ImgTex2[DialogCount], anchor = "nw")
            dialogo=canvas1.create_text( 30, 640, fill="Black",
                        text=dialogo2[DialogCount], anchor="nw",
                        font=("Helvetica",16))
        
        
def EncontrarPista():
    global dialogo
    global EnLugar
    global DialogCount
    global Imagen
    dialogo1=["Haku:\nMuy bien, chicos, dividanse y revisen \nsi encuentran algo.", 
             "No face: \nCaray!, encontré algo.",
             "No face:\nEncontre "+Malvado[2]+", \npero aun no se de quien es.",
             "BB:\nTal vez quien tenga a los padres de chihiro esté tramando algo."]
    ImgTex1=[haku_Img,noface_Img,noface_Img,bb_Img]
    
    dialogo2=["Haku:\nMuy bien, chicos, dividanse y encuentren alguna pista.",  
             "...",
             "Chihiro: \nAl parecer no hay una pista que nos ayude a encontrar a Scooby.",
             "No face: \nDebemos seguir investigando, pero tal vez no aqui."]
    ImgTex2=[haku_Img,BarraDialogo,chihiro_Img,noface_Img]
    
    if Malvado[2]==Mapa[i_mapa][2]: 
        if DialogCount>0 and DialogCount<len(dialogo1):  
            canvas1.itemconfig(Imagen,image =ImgTex1[DialogCount])
            canvas1.itemconfig(dialogo,text=dialogo1[DialogCount])
        
        elif DialogCount>=len(dialogo1): 
            
            DialogCount=0
            EnLugar=3
            canvas1.delete(dialogo)
            canvas1.delete(Imagen)
            EncontrarCollar()
            
        
        else:             
            Imagen=canvas1.create_image( 0, 0, image = ImgTex1[DialogCount], anchor = "nw")
            dialogo=canvas1.create_text( 30, 640, fill="Black",
                        text=dialogo1[DialogCount], anchor="nw",
                        font=("Helvetica",16))
    else:   
        if DialogCount>0 and DialogCount<len(dialogo2):  
            canvas1.itemconfig(Imagen,image =ImgTex2[DialogCount])
            canvas1.itemconfig(dialogo,text=dialogo2[DialogCount])
        
        elif DialogCount>=len(dialogo2): #
            DialogCount=0
            EnLugar=3
            canvas1.delete(dialogo)
            canvas1.delete(Imagen)
            EncontrarCollar()
            
        
        else:             
            Imagen=canvas1.create_image( 0, 0, image = ImgTex2[DialogCount], anchor = "nw")
            dialogo=canvas1.create_text( 30, 640, fill="Black",
                        text=dialogo2[DialogCount], anchor="nw",
                        font=("Helvetica",16))
    

def ubicar_mapa(lugar): 
    for i in range(5):
        if lugar==Mapa[i][1]:
            
            return i        
            break

def volver_menu():
    global dialogo
    global DialogCount
    global EnLugar
    EnLugar=0
    if AccionesCount==5:
        canvas1.delete(dialogo)
        DialogCount=0
        EnLugar=5
        
        ResolverMisterio()
            
    else:
        canvas1.pack(fill = "both", expand = True) 
        canvas1.pack() 
        canvas1.create_image( 0, 0, image = bg, anchor = "nw") 
        canvas1.create_image(0,0,image=BarraMenu,anchor="nw")
        mostrar_botones()
        boton_menu.place(x=5,y=20)
        canvas1.create_text( 400, 90, 
                        text="Selecciona un lugar para investigar ", fill="Black",
                        font=("Helvetica",18)) 
        canvas1.create_text( 400, 120, 
                        text=f"Intervenciones restantes: {5-AccionesCount}", fill="Black",
                        font=("Helvetica",18)) 

def siguiente():
    global DialogCount
    global AnswerBien
    DialogCount+=1
    if EnLugar==1:
        Destino(nom_lugar)
    elif EnLugar==2:
        EncontrarPista()
    elif EnLugar==3: 
        EncontrarCollar()
    elif EnLugar==4: 
        HablarSospechoso()
    elif EnLugar==5: 
        if DialogCount==1:
            
            siguiente.place_forget()
            mostrar_opciones()
        ResolverMisterio()
    elif EnLugar==6: 
        Final(AnswerBien)
    elif EnLugar==7:
        root.destroy()
        

def ResolverMisterio():
    global dialogo
    global DialogCount
    global ans
    global AnswerBien
    global EnLugar
    global Imagen
    
    DialogFinal=["No face:\nMuy bien, es hora de encontrar al secuestrador",
                "No face: \nQuien es el culpable del secuestro de los padres de chihiro?",
                "No face: \nEn donde secuestraron a los padres de chihiro?",
                "No face: \nFinalmente, Con qué objetivo?"]
    if DialogCount>0 and DialogCount<len(DialogFinal):  
        canvas1.itemconfig(dialogo,text=DialogFinal[DialogCount])
        
        boton_uno.configure(text=Mapa[0][DialogCount-1])
        boton_dos.configure(text=Mapa[1][DialogCount-1])
        boton_tres.configure(text=Mapa[2][DialogCount-1])
        boton_cuatro.configure(text=Mapa[3][DialogCount-1])
        boton_cinco.configure(text=Mapa[4][DialogCount-1])

    elif DialogCount>=len(DialogFinal): 
        DialogCount=0
        EnLugar=6
        canvas1.delete(dialogo)
        canvas1.delete(Imagen)
        ans=0
        for a in 'Trozo de carbon','Garra Mojada','Trozo de oro','Raton','Hilo suelto':
            if Malvado[2]==a:
                break
            ans+=1
        AnswerBien=True
        for i in range(3):
            if Malvado[i]!=Conclusion[i]:
                AnswerBien=False
       
        boton_uno.place_forget()
        boton_dos.place_forget()
        boton_tres.place_forget()
        boton_cuatro.place_forget()
        boton_cinco.place_forget()
        siguiente.place(x=700,y=550) 
        Final(AnswerBien)
        
    else:            
        canvas1.create_image( 0, 0, image = bg, anchor = "nw")
        Imagen=canvas1.create_image(0,0,image=noface_Img,anchor="nw")
        dialogo=canvas1.create_text( 30, 560, fill="",
                    text=DialogFinal[DialogCount], anchor="nw",
                    font=("Helvetica",16))
        
def resultado(respuesta):
    global DialogCount
    global Conclusion
    
    Conclusion.append(respuesta)
   
    DialogCount+=1  
    ResolverMisterio()

def Final(r):
    global ans
    global DialogCount
    global dialogo
    global Imagen, EnLugar
    if r==True: 
        objetivos=[f"las {Malvado[2]} que encontramos como pista? \nBueno, resulta que este carbon solo se encuentra\n en las calderas",
                  f"las {Malvado[2]} que encontramos como pista? \nBueno, resulta que esta garra pertenece a lin ",
                  f"los {Malvado[2]} que encontramos como pista? \nBueno, resulta que el trozo de oro \nEs lo que mas anhela el sapo",
                  f"los {Malvado[2]} que encontramos como pista? \nBueno, analicé detalladamente ese hilo suelto y es de zeniba",
                  f"los {Malvado[2]} que encontramos como pista? \nBueno, el raton corresponde con yubaba, \nen eso transformo a su ayudante"]
        
        Dialogo_Final=["No face: \nMuy bien,resuelto el caso. \nLa persona que secuestro a los padres de chihiro fue...",
                     f"Chihiro: \n{Malvado[0]} ??!!",
                     f"No face: \nAsi es, {Malvado[0]} secuestró a tus padres",
                     "BB: \nClaro, cuando fuimos a hablar, nos intentó engañar \nculpando a otras personas",
                     "Haku: \nHaciendo que deje de ser un sospechoso y poder seguir con su plan",
                     "Chihiro: \nY cual era su plan?",
                     f"No face: \nRecuerdas {objetivos[ans]}",
                      "Chihiro: \nMuy bien, se pudo resolver eso, pero donde están mis padres?",
                      f"{Malvado[0]}: \nNunca les diré donde se encuentran sus padr..",
                      "Hijaaaaaa",
                      "Chihiro: \nPapá,Mamá ?!!",
                      "papas: \nHija!!!",
                      "Chihiro: \ncomo escapaeon? estaba muy preocupada por ustedes",
                      "papas \nEstaba encerrado en uno de los sotano, \npero un huesped nos vio y nos salvó",
                      "Haku: \nBueno,están de regreso con nosotros",
                       f"{Malvado[0]}: \nRayos, me hubiera salido con la mia y \n chihiro se hubiera quedado aqui para siempre",
                      "Papás: \nVamonos hija !!!"]
        ImgFinal=[noface_Img,chihiro_Img,noface_Img,bb_Img,haku_Img,chihiro_Img,noface_Img,chihiro_Img,Malvado[3],
                 papas_Img,chihiro_Img,papas_Img,chihiro_Img,papas_Img,haku_Img,Malvado[3],papas_Img]
        
        if DialogCount>0 and DialogCount<len(Dialogo_Final): 
            canvas1.itemconfig(Imagen,image=ImgFinal[DialogCount])
            canvas1.itemconfig(dialogo,text=Dialogo_Final[DialogCount])
        
        elif DialogCount>=len(Dialogo_Final): 
            DialogCount=0
            EnLugar=0
            canvas1.delete(dialogo)
            canvas1.delete(Imagen)
            siguiente.place_forget()
           
            root.destroy()
        
        else:            
            Imagen=canvas1.create_image(0,0,image=ImgFinal[DialogCount],anchor="nw")
            dialogo=canvas1.create_text( 30, 640, fill="Black",
                        text=Dialogo_Final[DialogCount], anchor="nw",
                        font=("Helvetica",16))
    else:
        Imagen=canvas1.create_image(0,0,image=noface_Img,anchor="nw")
        dialogo=canvas1.create_text( 30, 640, fill="Black",
                        text=f"No face:Creo que aun te falta aprender a resolver crimenes. \nCulpable:{Malvado[0]} \nLugar de secuestro:{Malvado[1]} \nPista del objetivo:{Malvado[2]}", anchor="nw",
                        font=("Helvetica",16)) 
        EnLugar=7
#---------------------------------------------------------------------------  BOTONES DEL JUEGO
Oficina_Yubaba = Button(canvas1, text="Oficina Yubaba", width=12, command=cambia_oficina_yubaba)  #Cuando se presione un boton, se activará su metodo
Oficina_Yubaba.place(x=200,y=255)

Dormitorios = Button(canvas1, text="Dormitorios", width=12, command=cambia_Dormitorios)
Dormitorios.place(x=200,y=400)

Kamshi = Button(canvas1, text="Trabajo Kamshi", width=12, command=cambia_Kamshi)
Kamshi.place(x=220,y=550)

casa_zeniba = Button(canvas1, text="Casa Zeniba", width=12, command=cambia_casa_zeniba)
casa_zeniba.place(x=550,y=315)

puente = Button(canvas1, text="Puente", width=12, command=cambia_puente)
puente.place(x=400,y=550)

boton_menu = Button(canvas1, text="menu", width=12, command=volver_menu)
boton_menu.place(x=5,y=20)
siguiente = Button(canvas1, text="siguiente", width=12, command=siguiente)
siguiente.place(x=700,y=20)

boton_preguntar = Button(canvas1,text="Hablar con la persona",width="20",command=conversar,font=("Helveltica",12))
boton_investigar= Button(canvas1,text="Inspeccionar el lugar",width="20",command=investigar,font=("Helveltica",12))

boton_uno=Button(canvas1,text="",width="28",command= lambda:resultado(Mapa[0][DialogCount-1]),font=("Helveltica",12))
boton_dos=Button(canvas1,text="",width="28",command=lambda:resultado(Mapa[1][DialogCount-1]),font=("Helveltica",12))
boton_tres=Button(canvas1,text="",width="28",command=lambda:resultado(Mapa[2][DialogCount-1]),font=("Helveltica",12))
boton_cuatro=Button(canvas1,text="",width="28",command=lambda:resultado(Mapa[3][DialogCount-1]),font=("Helveltica",12))
boton_cinco=Button(canvas1,text="",width="28",command=lambda:resultado(Mapa[4][DialogCount-1]),font=("Helveltica",12))

root.mainloop() 
```
